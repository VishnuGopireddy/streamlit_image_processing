```python
def threshold_gray(img, threshold):
    '''
    :param img: grey sclae image as numpy array
    :param threshold: int threshold value
    :return: Binary gray scale image
    '''
    ret, binary = cv2.threshold(img,127, 255,0)
    return binary
```
